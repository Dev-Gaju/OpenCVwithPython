```python
#BLur and Smooth Image
#its need to detect the noise
```

```python
# Averaging
"""if you have 3*3 matrix and you wanna blur one matirx then the value will be average all the matrix."""
```

```python
average = cv.blur(img, (7,7))
cv.imshow('Average blur', average)
cv.imshow('Main image', img)
cv.waitKey(0)
```

```python
# Gaussian Blur
"""Not calculate the pixel value, it avarage the particular weight that given in each pixel then give the average in center"""
# Less blur compare to Avaeage blur and its called natural.
```

```python
guessian = cv.GaussianBlur(img, (3,3),0)  #0 represent standarad deviationinto x direction
cv.imshow('Average blur', average)
cv.imshow('Gaussian Blur ', guessian)
cv.waitKey(0)
```

```python
#median blur
""" same as average blur except find the average of sorrounding pixel it find the median"""
# More effective to reduce noise even better average and Gaussian blur 
```

```python
median = cv.medianBlur(img, 7)
cv.imshow('Median blur', median)
cv.imshow('Gaussian Blur ', guessian)
cv.waitKey(0)
```

```python
# Bilatral Bluring
# more effective 
#other blur image that effect edges but it not affect any edges
```

```python
bilateral =cv.bilateralFilter(img, 10, 35, 15)   #more give value look like median
cv.imshow('Bilateral blur', bilateral)
cv.imshow('Gaussian Blur ', guessian)
cv.waitKey(0)
```

```python
# BITWISE operations
```

```python
#4 types of bitwise operator
# and, or, xor, not
```

```python
blank = np.zeros((400,400), dtype='uint8')
```

```python
rectangle =cv.rectangle(blank.copy(),(30,30),(370,370),255,-1)  #image, margin,lest area, color, thicness
circle =cv.circle(blank.copy(), (200,200), 200,255,-1)
cv.imshow('Rectangle', rectangle)
cv.waitKey(0)
cv.imshow('Rectangle', circle)
cv.waitKey(0)
```

```python
# Bitwise AND
bitwise_and =cv.bitwise_and(rectangle,circle)   #retrun the intersection
cv.imshow('Bitwise AND', bitwise_and)    
cv.waitKey(0)
```

```python
# Bitwise OR
bitwise_or =cv.bitwise_or(rectangle,circle)   #retrun the intersection and not intersecting also
cv.imshow('Bitwise OR', bitwise_or)    
cv.waitKey(0)
```

```python
# Bitwise Xor
bitwise_xor =cv.bitwise_xor(rectangle,circle)   #Non intersecting mean return those area which not same on another
cv.imshow('Bitwise XOR', bitwise_xor)    
cv.waitKey(0)
```

```python
#getting and from subtracting
a =bitwise_or - bitwise_xor
cv.imshow('Bitwise And', a)
cv.waitKey(0)
```

```python
# bitwise Not
bitwise_not =cv.bitwise_not(circle)  #just invert the color that mean white invert to black and black unvert to white
cv.imshow("Bitwise Not",bitwise_not)
cv.waitKey(0)
```

```python
# !--Thresholding/Binarizing Images--!
# Thresholding is binarization of an image (0----255)(black---white)
#take some pixel value and thresold value by any chance pixel value less than threshold then we aassume puxel value as 0
```

```python
img = cv.imread('Photos/cat4.jpeg')
cv.imshow("Image yay!", img)
cv.waitKey(0)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow("Image GRAY!", gray)
cv.waitKey(0)
```

```python
# Simple Thresholding
thresold, thres = cv.threshold(gray, 230, 255, cv.THRESH_BINARY)#max and min value
#threshold return same value we pass and thress return  binarized value
cv.imshow("Simple Threshold", thres)
cv.waitKey(0)
```

```python
# threshold_invers
thresold, thres_invers = cv.threshold(gray, 230, 255, cv.THRESH_BINARY_INV)#max and min value
#threshold return same value we pass and thress return  binarized value
cv.imshow("Simple Threshold Invers", thres_invers)
cv.waitKey(0)
```

```python
# Adopted Thresholding  ---Specific threshold in several block
adavtive_threshold = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY_INV, 11,1) 
#block/kernel size 11 and c value is integer that subtracted from the mean
cv.imshow("adative Thresholding", adavtive_threshold)
cv.waitKey(0)
```

```python
#Edges in Images
#image have never negetive pixel value thats why used absolate value
```

```python
# Lablacian
lap = cv.Laplacian(gray, cv.CV_64F)  #data depth
lap = np.uint8(np.absolute(lap))
cv.imshow("laplecian", lap)
cv.waitKey(0)
```

```python
# Sobel more used than laplecian
sobelx = cv.Sobel(gray, cv.CV_64F,1,0)  # 1 and 0 means direction
sobely = cv.Sobel(gray, cv.CV_64F,0,1)
combines =cv.bitwise_and(sobelx,sobely)
cv.imshow("Sobel X", sobelx)
cv.imshow("Sobel Y", sobely)
cv.imshow("Comine Sobel ", combines)
cv.waitKey(0)
```

```python
#canny use most of the cases
canny  = cv.canny(gray, 150, 175)
cv.imshow("canny Images", canny)
cv.waitKey(0)
```
